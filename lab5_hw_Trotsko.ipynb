## 1 Логнормальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.lognorm.html).

```python
from scipy.stats import lognorm
```

Випадкова велиина $Y$ має логнормальний розподіл, якщо $Y = \exp \left\{ X \right\}$, де $X$ - випадкова величина з нормального розподілу.

```python
mu = 0
sigma = 1
```

Параметри $s, loc, scale$ відповідають логнормальному розподілу зі щільністю $$f(x) = \frac{1}{s (x - loc) \sqrt{2 \pi}} \exp \left\{ -\frac{1}{2 s^2} \log^2 \left( \frac{x - loc}{scale} \right) \right\}.$$

```python
s = sigma
loc = 0
scale = np.exp(mu)
```

### 1.1 Щільність

```python
x1 = np.arange(-1, 5, 0.1)
```

Виведіть значення функції щільності логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x1$.

```python
print(lognorm.pdf(x1, s, loc, scale))
```

Зобразіть графічно значення функції щільності логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x2$.

```python
x2 = np.arange(-1, 5.01, 0.01)
```

```python
fig, axs = plt.subplots(1,1)#, figsize=(20,5))

lognorm_den = lognorm.pdf(x2, s, loc, scale)
axs.plot(x2, lognorm_den, 'r')
axs.set_title('Lognorm pdf')

plt.show()
```

### 1.2 Функція розподілу

Виведіть значення функції розподілу логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x1$.

```python
print(lognorm.cdf(x1, s, loc, scale))
```

Зобразіть графічно значення функції розподілу логнормального розподілу з параметрами $s, loc, scale$ в точках вектора $x2$.

```python
fig, axs = plt.subplots(1,1)#, figsize=(20,5))

lognorm_distr = lognorm.cdf(x2, s, loc, scale)
axs.plot(x2, lognorm_distr, 'r')
axs.set_title('Lognorm cdf', color='b')

plt.show()
```

### 1.3 Випадкові величини

Згенеруйте $100$ випадкових величин з логнормального розподілу з параметрами $s, loc, scale$.

```python
print(lognorm.rvs(s, loc, scale, size=100))
```

Зобразіть гістограму $10000$ випадкових величин з логнормального розподілу з параметрами $s, loc, scale$.

```python
fig, axs = plt.subplots(1,1)

lognorm_rvs = lognorm.rvs(s, loc, scale, size=10000)
axs.hist(lognorm_rvs, bins = 200)
axs.set_title('Lognorm rvs', color='b')
axs.set_xlim(-1, 16)

plt.show()
```

Зобразіть графік функції щільності та нормовану гістограму $100000$ випадкових величин з логнормального розподілу з параметрами $s, loc, scale$. Поексперементуйте з аргументом $bins$ функції `hist`.

```python
fig, axs = plt.subplots(1,1)#, figsize=(20,5))

x = np.arange(-1, 10, 0.01)
lognorm_den = lognorm.pdf(x, s, loc, scale)
axs.plot(x, lognorm_den, 'r')
axs.hist(lognorm.rvs(s, loc, scale, size=100000), normed=True, bins=200)
axs.set_xlim(-1, 14)

plt.show()
```

## 2 Бета розподіл

```python
x2 = np.arange(-1, 2.01, 0.01)
```

```python
fig, axs = plt.subplots(1,1)#, figsize=(20,5))

beta_den = beta.pdf(x2, a, b, loc, scale)
axs.plot(x2, beta_den, 'r')

plt.show()
```

## 4 Апроксимація біноміального розподілу нормальним розподілом

Покажіть схожість біноміального розподілу та нормального розподілу з відповідними параметрами:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

Зобразіть графіки їх функцій щільності (обидві функції на одному графіку) та функцій розподілу (обидві функції на другому графіку) на двох сабплотах в одному рядку.

**Зауваження.** Математичне сподівання та дисперсія відповідного нормального розподілу збігається з математичним сподіванням та дисперсією біноміального розподілу.

```python
from scipy.stats import norm
from scipy.stats import binom


n = 100
prob = [0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]

for p in prob:
    fig, axs = plt.subplots(1,2, figsize=(20,5))
    loc = n * p
    scale = (loc * (1 - p)) ** 0.5
    
    x = np.arange(loc - 20, loc + 20, 0.01)
    x_bin = np.arange(loc - 20, loc + 20)

    axs[0].plot(x, norm.pdf(x, loc, scale), color='r')
    axs[0].stem(x_bin, binom.pmf(x_bin, n, p) )
    axs[0].set_title('pdf (n = {}, p = {}), (blue - binomial / red - normal)'.format(n, p))
    axs[0].set_ylim(0, 0.2)
    
    axs[1].plot(x, norm.cdf(x, loc, scale), 'r')
    axs[1].plot(x, binom.cdf(x, n, p), 'b')
    axs[1].set_title('cdf (n = {}, p = {}), (blue - binomial / red - normal)'.format(n, p))

    plt.show()
```

```python
def sort_dots(a, w, z):
    blue = []
    red = []
    for i in range(len(w)):
        if w[i] ** 2 + z[i] ** 2 > a ** 2:
            blue.append((w[i], z[i]))
        else:
            red.append((w[i], z[i]))
    return list(zip(*blue)), list(zip(*red))
    

param = [10, 100, 500, 1000, 2000, 10000]
a = 5

for n in param:
    dots_x = [_ for _ in uniform.rvs(loc=-a, scale=2*a, size=n)]
    dots_y = [_ for _ in uniform.rvs(loc=-a, scale=2*a, size=n)]
    b, r = sort_dots(a, dots_x, dots_y)
    # plotting dots
    fig, axs = plt.subplots(1, 1, figsize=(7, 7))
    axs.plot(*b, 'bo')
    axs.plot(*r, 'ro')
    
    # drawing a circle of radius a
    theta = np.arange(0, 2 * np.pi, 0.01)
    axs.plot(a * np.sin(theta), a * np.cos(theta), color='r', lw=5) 
    # drawing a square with a side of length a
    x = np.arange(-a, a, 0.01)
    y1 = [a] * len(x)
    y2 = [-a] * len(x)
    axs.plot(x, y1, x, y2, y1, x, y2, x, color='b', lw=4)
    
    axs.set_title('n = {}, pi = {}'.format(n, 4 * len(r[0]) / n))
    axs.set_xlim(-a - 0.5, a + 0.5)
    axs.set_ylim(-a - 0.5, a + 0.5)
    
    plt.show()
```
