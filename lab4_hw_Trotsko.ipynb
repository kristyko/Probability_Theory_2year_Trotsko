## 1 Від'ємний біноміальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.nbinom.html).

```python
from scipy.stats import nbinom
```

```python
n = 5
p = 0.4
```

### 1.1 Щільність

```python
x = np.arange(-0.5, n*5+1, 0.5)
```

Виведіть значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
den = nbinom.pmf(x, n, p)
print(den)
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.stem(x, den)
plt.show()
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x1$.

```python
x1 = np.arange(0, n*5+1)
```

```python
den1 = nbinom.pmf(x1, n, p)
plt.stem(x1, den1)
plt.show()
```

### 1.2 Функція розподілу

Виведіть значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
f = nbinom.cdf(x, n, p)
print(f)
```

Зобразіть графічно значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.plot(x, f, drawstyle='steps-post')
plt.show()
```

### 1.3 Випадкові величини

Згенеруйте $100$ випадкових величин з від'ємного біноміального розподілу з параметрами $n, p$.

```python
N = 100
r = nbinom.rvs(n, p, size=N)
# print(r)
```

Виведіть кількість кожного елемента в отриманому випадковому векторі.

```python
for i in x1:
    summ = sum(r==i)
    print('Кількість {} : {}, {}'.format(i, summ, summ / N))
```

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

```python
plt.hist(r)
plt.show()
```

## 3 Різні параметри

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom


param = [(15, 0.8), (15, 0.2), (20, 0.4), (20, 0.6), (7, 0.5), (25, 0.5), (100, 0.5)]
for n, p in param:
    x = np.arange(0, n+1)
    plt.subplot(1, 3, 1)
    den = binom.pmf(x, n, p)
    plt.stem(x, den)
    
    plt.subplot(1, 3, 2)
    f = binom.cdf(x, n, p)
    plt.plot(x, f, drawstyle='steps-post')
    plt.title('n = {}, p = {}'.format(n, p))
    
    plt.subplot(1, 3, 3)
    r = binom.rvs(n, p, size=10000)
    plt.hist(r)
    plt.show()
```

### 3.2 Геометричний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин геометричного розподілу з параметрами:
1. $p = 0.1$
2. $p = 0.2$
3. $p = 0.3$
4. $p = 0.4$
5. $p = 0.5$
6. $p = 0.6$
7. $p = 0.7$
8. $p = 0.8$
9. $p = 0.9$

```python
from scipy.stats import geom

param = np.arange(0.1, 1, 0.1)
for p in param:
    x = np.arange(0, 11)
    
    plt.subplot(1, 3, 1)
    den = geom.pmf(x, p)
    plt.stem(x, den)
    
    plt.subplot(1, 3, 2)
    f = geom.cdf(x, p)
    plt.plot(x, f, drawstyle='steps-post')
    plt.title('p = {}'.format(p))
    
    plt.subplot(1, 3, 3)
    r = geom.rvs(p, size=10000)
    plt.hist(r)
    plt.show()
```

### 3.3 Розподіл Пуассона

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин розподілу Пуассона з параметрами:
1. $\mu = 0.1$
2. $\mu = 0.2$
3. $\mu = 0.3$
4. $\mu = 0.4$
5. $\mu = 0.5$
6. $\mu = 0.6$
7. $\mu = 0.7$
8. $\mu = 0.8$
9. $\mu = 0.9$

```python
from scipy.stats import poisson

param = np.arange(0.1, 1, 0.1)
for mu in param:
    x = np.arange(0, 11)
    
    plt.subplot(1, 3, 1)
    den = poisson.pmf(x, mu)
    plt.stem(x, den)
    
    plt.subplot(1, 3, 2)
    f = poisson.cdf(x, mu)
    plt.plot(x, f, drawstyle='steps-post')
    plt.title('mu = {}'.format(mu))
    
    plt.subplot(1, 3, 3)
    r = poisson.rvs(mu, size=10000)
    plt.hist(r)
    plt.show()
```

### 3.4 Від'ємний біноміальний розподіл

```python
param = [(15, 0.8), (15, 0.2), (20, 0.4), (20, 0.6), (7, 0.5), (25, 0.5), (100, 0.5)]
for n, p in param:
    x = np.arange(0, n*5+1)
    plt.subplot(1, 3, 1)
    den = nbinom.pmf(x, n, p)
    plt.stem(x, den)

    plt.subplot(1, 3, 2)
    f = nbinom.cdf(x, n, p)
    plt.plot(x, f, drawstyle='steps-post')
    plt.title('n = {}, p = {}'.format(n, p))
    
    plt.subplot(1, 3, 3)
    plt.hist(nbinom.rvs(n, p, size=10000))
    plt.show()
```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
n = 100
prob = [0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]

for p in prob:
    x = np.arange(0, n+1)
    d1 = binom.pmf(x, n, p)
    d2 = poisson.pmf(x, p*n)
    plt.stem(x, d1, 'b', markerfmt='bo', label='binom') 
    plt.stem(x, d2, 'r', markerfmt='ro', label='poisson')
    plt.legend()
    plt.show()
```
